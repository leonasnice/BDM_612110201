```python
number = input("")
eo = number.split(",")
even,odd = 0,0
for num in eo: 
    if int(num) % 2 == 0: 
        even += 1
    else: 
        odd += 1       
print("Even numbers in the list: ", even) 
print("Odd numbers in the list: ", odd)
```
